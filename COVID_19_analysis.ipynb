```python
def plot_map(df, col, color, title):
  fig = px.choropleth(df,
      locations="Country/Region",
      locationmode="country names",
      color=col,
      color_continuous_scale=color,
      title=title
  )

  fig.show()
```

```python
plot_map(df, "Population", "YlOrRd", "Population Worldwide")
```

```python
plot_map(df, "TotalCases", "matter", "COVID-19 Cases Worldwide")
```

```python
plot_map(df, "ActiveCases", "matter", "COVID-19 Active Cases Worldwide")
```

```python
plot_map(df, "Serious,Critical", "matter", "COVID-19 Serious/Critical Cases Worldwide")
```

```python
plot_map(df, "TotalDeaths", "matter", "COVID-19 Deaths Worldwide")
```

```python
plot_map(df, "TotalTests", "Viridis", "COVID-19 Tests Worldwide")
```

```python
plot_map(df, "TotalRecovered", "Viridis", "COVID-19 Recovered Worldwide")
```

```python
plot_map(df, "Tot Cases/1M pop", "matter", "COVID-19 Total Cases/1M Population Worldwide")
```

```python
plot_map(df, "Deaths/1M pop", "matter", "COVID-19 Deaths/1M Population Worldwide")
```

```python
plot_map(df, "Tests/1M pop", "Viridis", "COVID-19 Tests/1M Population Worldwide")
```

```python
def plot_scatter(temp, x, y, title):
  fig = px.scatter(
      temp,
      x = x,
      y = y,
      color="Continent",
      hover_name="Country/Region",
      log_x=True,
      log_y=True,
      title=title,
      color_discrete_sequence=px.colors.qualitative.Vivid
  )
  fig.update_traces(textposition="top center")
  fig.show()
```

```python
plot_scatter(df, "TotalCases", "TotalDeaths", "Cases vs Deaths")
```

```python
plot_scatter(df, "Tot Cases/1M pop", "Deaths/1M pop", "Cases vs Deaths per 1M pop")
```

```python
plot_scatter(df, "Population", "TotalCases", "Population vs Cases")
```

```python
plot_scatter(df, "Population", "TotalTests", "Population vs Tests")
```

```python
plot_scatter(df, "Population", "Tot Cases/1M pop", "Population vs Total Cases/1M population")
```

```python
plot_scatter(df, "Population", "Deaths/1M pop", "Population vs Deaths/1M population")
```

```python
plot_scatter(df, "TotalDeaths", "TotalRecovered", "Deaths vs Recovered")
```
