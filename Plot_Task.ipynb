```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("csd.csv")
df.plot.bar(x = 'month_number', y = ['facewash', 'facecream'], color = ['b','r'])
plt.xlabel("Month Number")
plt.ylabel("Sales unit in number")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("csd.csv")
facecream = sum(df.facecream)
facewash = sum(df.facewash)
toothpaste = sum(df.toothpaste)
bathingsoap = sum(df.bathingsoap)
shampoo = sum(df.shampoo)
moisturizer = sum(df.moisturizer)

data = [facecream,facewash,toothpaste,bathingsoap,shampoo,moisturizer]


plt.pie(data,autopct='%1.1f%%' , labels = ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]  )
plt.title("Sales Data")
plt.show()



plt.grid(color='k', linestyle = '--' ,linewidth=1, axis='both', alpha=0.5)
```
